```python
import pandas as pd
df=pd.read_csv("/environmental_data.csv")

import seaborn as sns
import matplotlib.pyplot as plt

# Line graph for temperature, humidity, and rainfall
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df.index, y="Temperature", label="Temperature")
sns.lineplot(data=df, x=df.index, y="Humidity", label="Humidity")
sns.lineplot(data=df, x=df.index, y="Pressure", label="Pressure")

plt.title("Line Graph of Temperature, Humidity, and Rainfall")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()

```

```python
#using bar plot to make the bar plot fine thegraphs
import seaborn as sns
import matplotlib.pyplot as plt

# Temperature Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df.index, y="Temperature", data=df, palette="coolwarm")
plt.title("Temperature Bar Plot")
plt.xlabel("Index")
plt.ylabel("Temperature")
plt.show()
```

```python
# Humidity Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df.index, y="Humidity", data=df, palette="Blues")
plt.title("Humidity Bar Plot")
plt.xlabel("Index")
plt.ylabel("Humidity")
plt.show()
```

```python
# Pressure Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=df.index, y="Pressure", data=df, palette="Greens")
plt.title("Pressure Bar Plot")
plt.xlabel("Index")
plt.ylabel("Pressure")
plt.show()
```

```python

#using matplotlib and seaborn to make the pie graphs
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # For colormaps
import numpy as np

# Ensure all values are non-negative and handle NaN
df["Temperature"] = df["Temperature"].clip(lower=0).fillna(0)
df["Humidity"] = df["Humidity"].clip(lower=0).fillna(0)
df["Pressure"] = df["Pressure"].clip(lower=0).fillna(0)

# Labels for pie chart (e.g., use the index)
labels = df.index.astype(str)

# Temperature Pie Chart
plt.figure(figsize=(6, 6))
colors = cm.coolwarm(np.linspace(0, 1, len(df)))  # Generate colors from colormap
plt.pie(df["Temperature"], labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Temperature Pie Chart")
plt.show()

# Humidity Pie Chart
plt.figure(figsize=(6, 6))
colors = cm.Blues(np.linspace(0, 1, len(df)))
plt.pie(df["Humidity"], labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Humidity Pie Chart")
plt.show()

# Pressure Pie Chart
plt.figure(figsize=(6, 6))
colors = cm.Greens(np.linspace(0, 1, len(df)))
plt.pie(df["Pressure"], labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Pressure Pie Chart")
plt.show()
```
